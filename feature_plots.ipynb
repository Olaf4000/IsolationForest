```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
l3T = pd.read_csv('data/l3T.csv')
l3T.head()
```

# Plotting feature combinations

## clientIp / time_active

```python
plt.figure(figsize=(15, 7))

plt.scatter(l3T['clientIp'], l3T['time_active'])

plt.xticks(rotation=90)
#plt.axvline(x=10, color='black', linestyle='--')
plt.show()
```

### visualising different hosts

```python
plt.figure(figsize=(15, 7))

for host in l3T['host'].unique():
    l3F1 = l3T[l3T['host'] == host]
    plt.scatter(l3F1['clientIp'], l3F1['time_active'], label=host)

plt.xticks(rotation=90)
#plt.axvline(x=10, color='black', linestyle='--')
plt.legend()
plt.show()
```

## amount of requests / time active

```python
plt.figure(figsize=(15, 7))

plt.scatter(l3T['time_active'], l3T['count'])

plt.xticks(rotation=90)
#plt.axvline(x=10, color='black', linestyle='--')
plt.show()
```

### visualising different clientIps

```python
plt.figure(figsize=(15, 7))

for client in l3T['clientIp'].unique():
    l3F1 = l3T[l3T['clientIp'] == client]
    plt.scatter(l3F1['time_active'], l3F1['count'])

plt.xticks(rotation=90)
#plt.axvline(x=10, color='black', linestyle='--')
plt.xlabel('time_active')
plt.ylabel('count')
plt.show()
```
